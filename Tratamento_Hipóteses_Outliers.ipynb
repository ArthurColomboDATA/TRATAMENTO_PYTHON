## Importando módulos e dados

```python
#import de módulos
import pandas as pd
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import de dados
df_2019 = pd.read_csv("/content/vendas_linha_petshop_2019.csv", delimiter=";", encoding='ISO-8859-1')
df_2020 = pd.read_csv("/content/vendas_linha_petshop_2020.csv", delimiter=";", encoding='ISO-8859-1')
df_2021 = pd.read_csv("/content/vendas_linha_petshop_2021.csv", delimiter=";", encoding='ISO-8859-1')
df_2022 = pd.read_csv("/content/vendas_linha_petshop_2022.csv", delimiter=";", encoding='ISO-8859-1')
```

```python
#concatenação dos dfs originados pela importação dos dados
df_geral = pd.concat([df_2019, df_2020, df_2021, df_2022])
```

Tipos dos Dados Atuais:

```python
#Observação do tipo de dados armazenado
df_geral.dtypes
```

### Transformando 'object' em 'float64'


Iremos realizar o ajuste no tipo de dados, transformando-os, quando necessário, de object para float64 \
No caso, as colunas que irão receber tal ajuste, serão: quantidade, valor_total_bruto, valor, valor_comissao e lucro_liquido \
Além disso, iremos substituir os valores NaN, por 0

```python
# Substitui valores vazios (' ') e NaN por 0
columns_to_replace = ['quantidade', 'valor_total_bruto', 'valor', 'valor_comissao', 'lucro_liquido']
df_geral[columns_to_replace] = df_geral[columns_to_replace].replace([' ', np.nan], 0)

# Substitui vírgula por ponto e converte para float
for col in columns_to_replace:
    df_geral[col] = df_geral[col].astype(str).str.replace(',', '.').astype(float)

# Converte a coluna de data para o tipo datetime
df_geral['data'] = pd.to_datetime(df_geral['data'], dayfirst=True)

# Verifica os tipos de dados
df_geral.dtypes
```

```python
df_geral.head(5)
```

Agora percebemos que os *Missings*, conhecidos como NaN, se transformaram em valores numéricos do tipo float64, ou seja, se transformaram em **'0.0'**

## Aplicando nosso cálculo para descobrir as quantidades corretas de valores MISSING

Foram identificados e ajustados os valores missings, que compreende as colunas quantidade, valor_total_bruto, valor, valor_comissao e lucro_liquido,\
sendo assim, iremos realizar o ajuste dos valores que necessitam. Ex.:\

Na coluna Quantidade, linha 3, a informação não foi registrada, logo aparecia como NaN, se a coluna 'valor_total_bruto' registra 44 reais, e a coluna 'valor' registra 22 reais, iremos realizar uma conta de divisão para capturar a quantidade vendida, sendo assim a quantidade correta referente a essa linha que estamos observando é '2.0' (44/22).\

Esse ajuste ocorrerá para todas as colunas listadas acima.

**Para todo o valor 0 armazenado na coluna 'quantidade', realizaremos uma divisão entre o 'valor_total_bruto' e a 'valor',
adicionando o resultado novamente na coluna 'qauntidade'; essa operação contemplará valores por linha.**

```python
df_geral['quantidade'] = np.where((df_geral['quantidade'] == 0),
                                 df_geral['valor_total_bruto'] / df_geral['valor'],
                                  df_geral['quantidade'])
```

**Para todo o valor 0 armazenado na coluna 'valor', realizaremos uma divisão entre o 'valor_total_bruto' e a 'quantidade',
adicionando o resultado novamente na coluna 'valor'; essa operação contemplará valores por linha.**

```python
df_geral['valor'] = np.where((df_geral['valor'] == 0),
                                 df_geral['valor_total_bruto'] / df_geral['quantidade'],
                                  df_geral['valor'])
```

**Para todo o valor 0 armazenado na coluna 'valor_total_bruto', realizaremos uma multiplicação entre o 'valor' e a 'quantidade',
adicionando o resultado novamente na coluna 'valor_total_bruto'; essa operação contemplará valores por linha.**

```python
df_geral['valor_total_bruto'] = np.where((df_geral['valor_total_bruto'] == 0),
                                 df_geral['valor'] * df_geral['quantidade'],
                                  df_geral['valor_total_bruto'])
```

Fizemos o nosso cálculo para todos os valores das colunas que estavam em '0.0' com os códigos acima, resultando em um novo DataFrame:

### Tratando os outliers

Vamos utilizar do metódo de Winsorizar, para tratar os outliers identificados, que consiste em limitar ou "capar" os valores extremos de uma distribuição de dados.

#### Para coluna 'valor':

```python
# Tratamento Valor
# 1º obtemos os quartis
wins_values_v = df_geral['valor'].quantile([0.05, 0.9]).to_list()

# Agora aplicamos os limites
df_geral['valor_wins'] = df_geral['valor'].clip(wins_values_v[0], wins_values_v[1])

# Por último vemos os dados como ficaram
df_geral[['valor', 'valor_wins']].head(10)
```

```python
#Box Plot da coluna 'valor'
df_geral[['valor', 'valor_wins']].boxplot()
```

Aqui podemos observar, graficamente, a diferença entre "valor" e "valor_wins". \
valor_wins demonstra que ainda temos outliers porem foram minimizados consideravelmente!

**Agora iremos fazer o mesmo método para as colunas restantes!**

#### Para coluna 'quantidade':

```python
# Tratamento Qntd
# obter os quartis
wins_values = df_geral['quantidade'].quantile([0.05, 0.9]).to_list()

# winsorizar
df_geral['qtd_wins'] = df_geral['quantidade'].clip(wins_values[0], wins_values[1])

# ver os dados
df_geral[['quantidade', 'qtd_wins']].head(10)
```

```python
#Box Plot da coluna 'quantidade'
df_geral[['quantidade', 'qtd_wins']].boxplot()
```

Diferente do 'valor', na coluna 'quantidade' pudemos zerar os outliers com este método!

## Calculando a matriz de correlação dos dados fornecidos

### Perguntas à serem respondidas:

### Deixando a correlaçao mais visível

```python
df_correlacao = df_geral.select_dtypes(include=['number']).corr()
df_correlacao
```

```python
plt.figure(figsize=(15, 12))
# Selecionando as colunas desejadas para a correlação
colunas = ['cod_pedido', 'valor_wins', 'qtd_wins', 'valor_tt_wins', 'lucro_wins', 'comissao_wins']
# Calculando a matriz de correlação apenas para essas colunas
df_correlacao_selecionada = df_correlacao[colunas].corr()

# Plotando o heatmap
sns.heatmap(df_correlacao_selecionada, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação')
plt.show()
```

### Resposta:

As variáveis que apresentam forte correlação positiva, são as tonalidade mais avermelhadas, sendo a maior correlação a comissão_wins e lucro_wins. 

Já as variáveis que apresentam forte correlação negativa, são as de tonalidade mais azuladas, sendo a menor correlação a qtd_wins e valor_wins.

**Insights:**
- O lucro_wins e comissao_wins estão diretamente relacionadas, o que indica que a comissão pode ser calculada como uma porcentagem do lucro. Isso faz sentido, pois, se usarmos a lógica, quanto maisa vendas, maior o lucro e maior a comissão.
- O valor_wins parcece ser um bom indicador para valor_total_wins, lucro_wins e comissao_wins.
- Na negativa, podemos apontar a correlação entre oferta e demanda, ou seja, uma correlação linear entre quantidade e valor, quanto maior quantidade daquele produto, menor o valor dele.
